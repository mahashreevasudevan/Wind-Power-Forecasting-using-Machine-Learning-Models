```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import shap
import matplotlib.pyplot as plt

# Load the data
file_path = 'D:/Wind/NEW PAPER/GUJARAT/JAFF DATA FIN.xlsx'  # Change the file path accordingly
data = pd.read_excel(file_path)

# Preprocessing
data['DATETIME'] = pd.to_datetime(data['DATETIME'])
data = data.sort_values(by='DATETIME')  # Ensure data is sorted by datetime

# Feature and target
X = data[['WS']].values  # Using wind speed as the input
y = data['Power'].values  # Wind power as the target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Define Kolmogorov-Arnold Network (KAN) Model
class KAN(tf.keras.Model):
    def __init__(self, num_layers=5, units_per_layer=64):
        super(KAN, self).__init__()
        self.num_layers = num_layers
        self.units_per_layer = units_per_layer

        # Hidden layers (Nonlinear)
        self.hidden_layers = []
        for _ in range(self.num_layers):
            self.hidden_layers.append(tf.keras.layers.Dense(self.units_per_layer, activation='tanh'))

        # Output layer
        self.output_layer = tf.keras.layers.Dense(1)

    def call(self, inputs):
        x = inputs
        for hidden_layer in self.hidden_layers:
            x = hidden_layer(x)
        return self.output_layer(x)

# Instantiate the KAN model
kan_model = KAN(num_layers=5, units_per_layer=64)

# Use the AdamW optimizer with weight decay
adamw_optimizer = tf.keras.optimizers.AdamW(learning_rate=0.001, weight_decay=1e-4)

# Compile the model with AdamW optimizer
kan_model.compile(optimizer=adamw_optimizer, loss='mse')

# Train the model and save training history
history = kan_model.fit(X_train, y_train, epochs=200, batch_size=32, verbose=1, validation_split=0.2)

# Forecasting using the trained model
y_pred = kan_model.predict(X_test)

# Calculate RMSE, MSE, and MAE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"RMSE: {rmse}")
print(f"MSE: {mse}")
print(f"MAE: {mae}")

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss Curve')
plt.legend()
plt.grid()
plt.show()

# SHAP Interpretability
explainer = shap.KernelExplainer(kan_model.predict, X_train)

# Calculate SHAP values for the first 10 samples
shap_values = explainer.shap_values(X_test[:10])  # Explain first 10 samples

# Plot SHAP values (Using the same subset of data for both SHAP and the plot)
shap.summary_plot(shap_values, X_test[:10])

# Plot actual vs forecasted wind power for the last 144 points
plt.figure(figsize=(12, 6))
plt.plot(data['DATETIME'][-144:], y_test[-144:], label="Actual Wind Power", marker='o')
plt.plot(data['DATETIME'][-144:], y_pred[-144:], label="Forecasted Wind Power", linestyle='--')
plt.xlabel("Date")
plt.ylabel("Wind Power(in MW)")
plt.title("Actual vs Forecasted Wind Power (Last 144 Points)")
plt.legend()
plt.grid()
plt.show()
```
